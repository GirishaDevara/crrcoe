```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Files/headbrain.csv")
data
#data.head()
```

```python
# Collecting X and Y
X = data['Head Size(cm^3)'].values
print(X)
Y = data['Brain Weight(grams)'].values
print(Y)
```

```python
# Calculating coefficient

# Mean X and Y
mean_x = np.mean(X)
print(mean_x)
mean_y = np.mean(Y)
print(mean_y)
```

```python
# Total number of values
n = len(X)

# Using the formula to calculate b1 and b2
numer = 0
denom = 0
for i in range(n):
    numer += (X[i] - mean_x) * (Y[i] - mean_y)
    denom += (X[i] - mean_x) ** 2
b1 = numer / denom
b0 = mean_y - (b1 * mean_x)
print(b1)
print(b0)
```

```python
# Printing coefficients
print("Coefficient, intercept")
print(b1, b0)   #where b1 stands for m and b0 stands for c
```

```python
# Plotting Values and Regression Line

max_x = np.max(X) 
min_x = np.min(X)

# Calculating line values x and y
x = np.linspace(min_x, max_x, 1000)
y = b0 + b1 * x

# Ploting Line
plt.plot(x, y, color='red', label='Regression Line')
# Ploting Scatter Points
plt.scatter(X, Y, c='blue', label='Scatter Plot')

plt.xlabel('Head Size in cm3')
plt.ylabel('Brain Weight in grams')
plt.legend()
plt.show()
```

```python
y_pred = b0 + b1*X[5]
print(y_pred)
# Calculating R2 Score
ss_tot = 0
ss_res = 0
for i in range(n):
    y_pred = b0 + b1 * X[i]
    ss_tot += (Y[i] - mean_y) ** 2
    ss_res += (y_pred - mean_y) ** 2
r2 = (ss_res/ss_tot)
print(y_pred)
print("R2 Score")
print(r2)
```

```python
# Model Intialization
reg = LinearRegression()
# Data Fitting
reg = reg.fit(X, Y)
# Y Prediction
Y_pred = reg.predict(X)
```
